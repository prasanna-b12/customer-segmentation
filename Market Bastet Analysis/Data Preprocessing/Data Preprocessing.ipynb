# Merging the data in One column

```python
#importing the library
import pandas as pd

#check whether header is there or not
data = pd.read_csv("csv/sample-1.csv", header=None)
data.head()
```

```python
#creating the dictionary to store the all products in one column
product_dict = {"products": []}

for x in range(len(data)):
    l = []
    
    #removing the column with null values
    products = data.iloc[x,:].dropna()
    for y in range(len(products)):
        pro = products[y]
        l.append(pro)
        
    #removing the square bracket from the list
    l_s = ", ".join(l)
    product_dict["products"].append(l_s)
```

```python
#converting to Dataframe
products_df = pd.DataFrame(product_dict)
#products_df.to_csv("D:\Datasets\MBA\products.csv", index=False)
products_df.head()
```

```python
#importing the library
import pandas as pd

#we can also get the  column name if we have many features
df = pd.read_csv("csv/products.csv")
df.head()
```

```python
pro = input("enter the col name: ")
```

```python
#creating the list to identify the unique products in the data
pro_list = []
for i in range(len(df)):
    pro_row = df[pro][i]
    d1= pro_row.split(",")
    for x in range(len(d1)):
        df1 = d1[x].strip()
        if df1 not in pro_list:
            pro_list.append(df1)
            
#print(pro_list)
```

```python
#creating the list to identify the unique products in the data
pro_list = []
for i in range(len(df)):
    pro_row = df['products'][i]
    d1= pro_row.split(",")
    for x in range(len(d1)):
        df1 = d1[x].strip()
        if df1 not in pro_list:
            pro_list.append(df1)
            
#print(pro_list)

#creating the dictionary to store whether they buy or not 1 or 0
dict_list = {}
for i in range(len(pro_list)):
    dict_list[pro_list[i]] = []
#print(dict_list)

#crerating the function to store the right value
def values(ch):
    for i in range(len(dict_list)):
        n = list(dict_list.keys())[i]
        if n in ch:
            dict_list[n].append(1)
        else:
            dict_list[n].append(0)

for i in range(len(df)):
    d9 = df['products'][i]
    d10= d9.split(",")
    #print(d10)
    li = []
    for x in range(len(d10)):
        df2 = d10[x].strip()
        li.append(df2)
    values(li)
```

```python
new_df = pd.DataFrame(dict_list)
new_df.head()
#new_df.to_csv("cleaned_data.csv", index= False)
```
